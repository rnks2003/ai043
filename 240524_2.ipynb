**Statistics :**

```python
from collections import Counter
import matplotlib.pyplot as plt
import random
```

```python
nFriends = []
for i in range(20):
    nFriends.append(random.randint(0,150))
```

```python
nFriends
```

```python
fCount = Counter(nFriends)
xs = range(151)
ys = [fCount[x] for x in xs]
```

```python
plt.bar(xs,ys)
plt.axis([0,151,0,20])
plt.title("Hist of fCount")
plt.xlabel("# of friends")
plt.ylabel("# of people")
plt.show()
```

**Mean, Median and Mode :**

```python
import statistics
```

```python
nums = [i for i in range(10)]

mean = statistics.mean(nums)
median = statistics.median(nums)
mode = statistics.mode(nums)
```

```python
print(f'Mean : {mean}\nMedian : {median}\nMode : {mode}')
```
