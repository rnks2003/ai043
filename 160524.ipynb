**A study was conducted to understand the effects of number of hours the student spent studying on their performance in the final exams.**
**Write a code to plot line chart with number of hours spent studying on x-axis and score in final exam on y-axis.**
**Use a red '*' as the point character, label the axes and give the plot a title**

*Importing dependencies :*

```python
import matplotlib.pyplot as plt
```

*Dataset pre-processing :*

```python
x = [10,9,2,15,10,16,11,16]
y = [95,80,10,50,45,98,38,93]

x,y = zip(*sorted(zip(x,y)))
```

*Plotting :*

```python
plt.plot(x,y)
plt.scatter(x,y,marker='*',color='red')
plt.xlabel("Number of hours spent studying (x)")
plt.ylabel("Score in final exam (y)")
plt.title('Effort vs Result')
```

*Dice roll :*

```python
import random

trials = 100000
sixes = 0

for i in range(trials):
    res = random.randint(1,6)
    if res ==6: sixes+=1

pSix = sixes/trials
```
