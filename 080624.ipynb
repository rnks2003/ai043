## Linear regression

**Implementation of LINEAR REGRESSION using least square estimator**

*Importing dependencies*

```python
import matplotlib.pyplot as plt
import numpy as np
```

*Helper method definition*

```python
def estimateCoef(x,y):
    n =np.size(x)
    mX = np.mean(x)
    mY = np.mean(y)
    sXY = np.sum(y*x) - n*mY*mX
    sXX = np.sum(x*x) - n*mX*mX
    
    b1 = sXY / sXX
    b0 = mY - b1*mX
    return (b0,b1)
```

```python
def plotRegLine(x,y,b):
    plt.scatter(x,y,color="b",marker="o",s=20,alpha=0.5)
    yHat = b[0] + b[1]*x
    plt.plot(x,yHat,color="r")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

*Runner :*

```python
np.random.seed(0)
x = np.random.rand(1000)
y = 20*x + 5*np.random.randn(len(x))
```

```python
b = estimateCoef(x,y)
print(f"Estimated Coefs :\nb0 = {b[0]}\nb1 = {b[1]}")
```

```python
plotRegLine(x,y,b)
```

**Using sklearn**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size=0.2,random_state=42)
linReg = LinearRegression()
```

```python
xTrain = xTrain.reshape(-1, 1)
yTrain = yTrain.reshape(-1, 1)
linReg.fit(xTrain,yTrain)
```

```python
linReg.coef_, linReg.intercept_
```

```python
yHat = linReg.predict(xTest.reshape(-1,1))
```

```python
plt.scatter(xTest,yTest,color='b',alpha=0.5)
plt.scatter(xTest,yHat,color='r',alpha=0.2)
```
