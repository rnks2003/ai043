**Implementation of LINEAR REGRESSION using least square estimator**

*Importing dependencies*

```python
import matplotlib.pyplot as plt
import numpy as np
```

*Helper method definition*

```python
def estimateCoef(x,y):
    n =np.size(x)
    mX = np.mean(x)
    mY = np.mean(y)
    sXY = np.sum(y*x) - n*mY*mX
    sXX = np.sum(x*x) - n*mX*mX
    
    b1 = sXY / sXX
    b0 = mY - b1*mX
    return (b0,b1)
```

```python
def plotRegLine(x,y,b):
    plt.scatter(x,y,color="b",marker="o",s=20,alpha=0.5)
    yHat = b[0] + b[1]*x
    plt.plot(x,yHat,color="r")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

*Runner :*

```python
x = np.arange(1000)
y = x/20 + 5*np.random.randn(len(x))
```

```python
b = estimateCoef(x,y)
print(f"Estimated Coefs :\nb0 = {b[0]}\nb1 = {b[1]}")
```

```python
plotRegLine(x,y,b)
```
