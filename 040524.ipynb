### 8 Arithmatics

```python
a = 25
b = 10

print("a+b = ",a+b)
print("a-b = ",a-b)
print("a*b = ",a*b)
print("a/b = ",a/b)
print("a//b = ",a//b)
print("a%b = ",a%b)

print("a>b = ",a>b)
print("a<b = ",a<b)

c = a+b
print("c = a+b = ",c)

print("3**2 = ",3**2)
print("a**b = ",a**b)

print("25>35 and 32<30 = ",25>35 and 32<30)
```

## Matplotlib Basics :

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting Linear curve :

*Plot y1 = x^2 and y2 = x^3*

```python
x = np.arange(10)
y1 = x**2
y2 = x**3

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["x^2","x^3"])
plt.show()
```

*Plot sin(x) and cos(x) functions*

```python
x = np.arange(360)
x = x*np.pi/180
sin = np.sin(x)
cos = np.cos(x)

plt.plot(x,sin)
plt.plot(x,cos)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["sin(x)","cos(x)"])
plt.show()
```

### Plotting bar graphs :

*Plot y=x^2*

```python
x = np.arange(10)
y = x**2

plt.bar(x,y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```

*Plot sin(x)*

```python
x = np.arange(100)
y = np.sin(x)

plt.bar(x,y)

plt.xlabel("x")
plt.ylabel("y = sin(x)")
plt.show()
```

### Plot scatter plot :

*Plot y1 = x^2 and y2 = x^3*

```python
x = np.arange(20)
y1 = x**2
y2 = x**3

plt.scatter(x,y1)
plt.scatter(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["x^2","x^3"])
plt.show()
```

*Plot sin(x) and cos(x)*

```python
x = np.arange(90)
x = x*np.pi/180
sin = np.sin(x)
cos = np.cos(x)

plt.scatter(x,sin)
plt.scatter(x,cos)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["sin(x)","cos(x)"])
plt.show()
```

*Labeled scatter plot*

```python
f = [70,65,72,63,71,64,60,64,67]
m = [175,170,205,120,220,130,105,145,190]
l = ['a','b','c','d','e','f','g','h','i']

plt.scatter(f,m)

for l, fCount, mCount in zip(l,f,m):
    plt.annotate(l, xy=(fCount,mCount), xytext=(5,-5),textcoords='offset points')

plt.title("Minutes per day VS Number of Friends")
plt.xlabel("Number of Friends")
plt.ylabel("Minutes per day spent on the site")
plt.show()
```
